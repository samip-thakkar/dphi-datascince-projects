```python
#Import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the train data
data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/kc_house_data/kc_house_data.csv')
data.head()
```

```python
#Get the test data
test_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/kc_house_data/kc_house_new_test_data.csv')
test_data.head()
```

```python
#Get the description of data
data.describe()
```

```python
#Get the info of data
data.info()
```

```python
#Divide the data into features and target
features = data.iloc[:, 1:]
target = data['price'].values
```

## Linear Regression using all features

```python
#Divide the data into train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, train_size = 0.8, random_state = 42)

#Perform Linear Regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model = lr.fit(X_train, y_train)
y_pred = model.predict(X_test)

#Evaluate the model
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_pred, y_test)
print("Root Mean squared error: ", mse ** 0.5)
```

```python
#Get the prediction on test data
y_pred = model.predict(test_data)
prediction = pd.DataFrame(y_pred, columns = ['prediction']).to_csv('results.csv', index = False)

#RMSE = 205089.08267176972
```

## Linear Regression using PCA

```python
#Normalize the data
mean_data = data.mean(axis = 0)
std_data = data.std(axis = 0)
print(mean_data.shape, std_data.shape)
```

```python
data_scaled = (data - mean_data[None, :]) / std_data[None, :]
data_scaled.head()
```

```python
features_scaled = data_scaled.iloc[:, 1:]
target_scaled = data_scaled['price']
features_scaled
```

```python
#Apply PCA
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
pca.fit(features_scaled)
features_transformed = pca.transform(features_scaled)
features_transformed.shape
```

```python
#Split in train and test
X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled = train_test_split(features_transformed, target_scaled, train_size = 0.8)
#Linear Regression
lr = LinearRegression()
model = lr.fit(X_train_scaled, y_train_scaled)
y_pred = model.predict(X_test_scaled)
mean_squared_error(y_pred, y_test_scaled)
```

```python
#Get the prediction on test data
test_data_transformed = pca.transform(test_data)
y_pred = model.predict(test_data_transformed)
prediction = pd.DataFrame(y_pred, columns = ['prediction']).to_csv('results.csv', index = False)

#RMSE: 666223
```

```python
features_selected = data[['bathrooms','sqft_living', 'grade', 'sqft_above', 'sqft_living15']]
target_selected = data['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_selected, target, train_size = 0.8, random_state = 42)

#Perform Linear Regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model = lr.fit(X_train, y_train)
y_pred = model.predict(X_test)

#Evaluate the model
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_pred, y_test)
print("Root Mean squared error: ", mse ** 0.5)
```

```python
#Test data 
test_data_features = test_data[['bathrooms','sqft_living', 'grade', 'sqft_above', 'sqft_living15']]
test_data_features.head()
```

```python
#Get the prediction on test data
y_pred = model.predict(test_data_features)
prediction = pd.DataFrame(y_pred, columns = ['prediction']).to_csv('results.csv', index = False)

#RMSE = 250000
```
