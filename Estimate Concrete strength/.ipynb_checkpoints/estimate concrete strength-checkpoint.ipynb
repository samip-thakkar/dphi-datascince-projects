#### Objective: The concrete compressive strength is a highly nonlinear function of age and ingredients. These ingredients include cement, blast furnace slag, fly ash, water, superplasticizer, coarse aggregate, and fine aggregate. Your objective is to build a machine learning model that would help Civil Engineers to estimate the compressive strength of the concrete and they can further take a decision whether the concrete should be used in their current project or not.

```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter('ignore')
```

```python
#Import the data
data = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/concrete_data/training_set_label.csv" )
data.head()
```

```python
#Load the test data
test_data  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/concrete_data/testing_set_label.csv")
test_data.head()
```

```python
#Get the description of data
data.describe()
```

```python
#Get the data info
data.info()
```

```python
#Check for null values
data.isna().sum()
```

```python
#Renaming the columns for simplicity
data.columns = ['Cement', 'Blast Furnace Slag', 'Fly Ash', 'Water', 'Superplasticizer', 'Coarse Aggregate', 'Fine Aggregate', 'Age', 'Concrete strength']
test_data.columns = ['Cement', 'Blast Furnace Slag', 'Fly Ash', 'Water', 'Superplasticizer', 'Coarse Aggregate', 'Fine Aggregate', 'Age']
data.head()
```

```python
#Divide the data into features and target variable
features = data.drop('Concrete strength', axis=1)
target = data['Concrete strength']

print(features.shape)
print(target.shape)
```

```python
#Divide the data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, train_size=0.75, random_state=42)

print(X_train.shape)
print(X_test.shape)
```

```python
# View the correlation among the features
plt.rcParams['figure.figsize'] = 12, 12
sns.heatmap(X_train.corr(), annot=True)
```

```python
#Train linear regression 
from sklearn.linear_model import LinearRegression
lr = LinearRegression(normalize=False)
lr_model = lr.fit(X_train, y_train)
```

```python
#Get the prediction
y_pred = lr_model.predict(X_test)
```

```python
#Calculate accuracy
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(y_pred, y_test) ** 0.5
print("RMSE: ", rmse)
```

```python
#Compare actual values and predictions
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, label="original")
plt.plot(x_ax, y_pred, label="predicted")
plt.legend()
plt.show()
```

```python
#Get the prediction on test data
y_hat = lr_model.predict(test_data)
#Save as csv
predictions = pd.DataFrame(y_hat, columns=['prediction'])
predictions.to_csv('results.csv', index=False, header=True)
```

### Standardize the data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_ = pd.DataFrame(sc.fit_transform(X_train), columns=features.columns)
X_test_ = pd.DataFrame(sc.transform(X_test), columns=features.columns)
test_data_ = pd.DataFrame(sc.transform(test_data), columns=features.columns)
test_data_
```

```python
#Apply PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X_train_)
X_train_reduced = pca.transform(X_train_)
X_test_reduced = pca.transform(X_test_)
test_data_reduced = pca.transform(test_data_)
X_train_reduced
```

```python
lr = LinearRegression()
lr_model = lr.fit(X_train_reduced, y_train)
y_pred = lr_model.predict(X_test_reduced)
```

```python
print(mean_squared_error(y_test, y_pred) ** 0.5) 
```

```python
#Compare actual values and predictions
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, label="original")
plt.plot(x_ax, y_pred, label="predicted")
plt.legend()
plt.show()
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf_model = rf.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
print(mean_squared_error(y_test, y_pred) ** 0.5) 
```

```python
#Compare actual values and predictions
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, label="original")
plt.plot(x_ax, y_pred, label="predicted")
plt.legend()
plt.show()
```

```python
#Hyper parameter tuning of RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
rf = RandomForestRegressor()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train, y_train)

rf_random.best_params_
```

```python
rf = RandomForestRegressor(n_estimators=1400, min_samples_split=2, min_samples_leaf=1, max_features='auto', max_depth=100, bootstrap=True)
rf_model = rf.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
print("RMSE: ", mean_squared_error(y_pred, y_test) ** 0.5)
```

```python
#Compare actual values and predictions
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, label="original")
plt.plot(x_ax, y_pred, label="predicted")
plt.legend()
plt.show()
```

```python
y_hat = rf_model.predict(test_data)
prediction = pd.DataFrame(y_hat, columns=['prediction']).to_csv('results.csv', index=False, header=True)
```

### Boosting Algorithms

```python
#XGBoost algorithm
import xgboost as xgb
xg = xgb.XGBRegressor()
xgb_model = xg.fit(X_train, y_train)
y_pred = xgb_model.predict(X_test)
print("RMSE: ", mean_squared_error(y_pred, y_test) ** 0.5)
```

```python
#Compare actual values and predictions
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, label="original")
plt.plot(x_ax, y_pred, label="predicted")
plt.legend()
plt.show()
```

```python
y_hat = xgb_model.predict(test_data)
prediction = pd.DataFrame(y_hat, columns=['prediction']).to_csv('results.csv', index=False, header=True)
```

```python
#Hyperparameter tuning XGBoostRegressor
params = {'colsample_bytree': 1.0,
 'eta': 0.01,
 'eval_metric': 'mae',
 'max_depth': 10,
 'min_child_weight': 6,
 'objective': 'reg:linear',
 'subsample': 0.8}

train_mat = xgb.DMatrix(X_train, label=y_train)
val_mat = xgb.DMatrix(X_test, label=y_test)

xgb_model = xgb.train(
    params,
    train_mat,
    num_boost_round=1000,
    evals=[(val_mat, "Test")],
    early_stopping_rounds=10
)
```

```python
y_pred = xgb_model.predict(val_mat)
print("RMSE: ", mean_squared_error(y_pred, y_test) ** 0.5)
```

```python
#Compare actual values and predictions
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, label="original")
plt.plot(x_ax, y_pred, label="predicted")
plt.legend()
plt.show()
```
